## Data Loading and Inspection

### Load in data from the relevant directory and inspect the dataframe.

```python
# Load the main dataset with the proper text encoding
df = pd.read_csv('data/AviationData.csv', encoding='latin-1')

# Take a peek at the first 5 rows to see what columns we have
df.head()
```

## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:

```python
# Inspect the total rows, column names, and missing values
df.info()

# Check out the basic summary statistics for numerical columns
df.describe()
```

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
# Convert Event.Date to datetime so we can extract the year
df['Event.Date'] = pd.to_datetime(df['Event.Date'])

# Filter for flights from 1983 onward AND where the aircraft is NOT amateur built
# We look for 'No' in the 'Amateur.Built' column to keep professional builds
cleaned_df = df[(df['Event.Date'].dt.year >= 1983) & (df['Amateur.Built'] == 'No')].copy()

# Check how many rows we have left after filtering
print(f"Rows after filtering: {len(cleaned_df)}")
```

**Aircraft.Damage**
- identify and execute any cleaning tasks
- construct a derived column tracking whether an aircraft was destroyed or not.

```python
# 1. Standardize the strings by stripping whitespace and making it Title Case
cleaned_df['Aircraft.damage'] = cleaned_df['Aircraft.damage'].str.strip().str.title()

# 2. Fill missing damage records with 'Unknown'
cleaned_df['Aircraft.damage'] = cleaned_df['Aircraft.damage'].fillna('Unknown')

# 3. Create a derived column tracking if the aircraft was completely destroyed
# This will return True if 'Destroyed', and False otherwise
cleaned_df['Is_Destroyed'] = cleaned_df['Aircraft.damage'] == 'Destroyed'

# Quick verification to see the distribution of values
print(cleaned_df['Aircraft.damage'].value_counts())
print("\nIs_Destroyed breakdown:")
print(cleaned_df['Is_Destroyed'].value_counts())
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
# 1. Clean up basic text spacing and force everything to uppercase
cleaned_df['Make'] = cleaned_df['Make'].str.strip().str.upper()

# 2. Consolidate a few of the top messy names using basic text replacement
cleaned_df['Make'] = cleaned_df['Make'].replace({
    'CESSNA AIRCRAFT CO': 'CESSNA', 'CESSNA AIRCRAFT': 'CESSNA',
    'PIPER AIRCRAFT CORP': 'PIPER', 'PIPER AIRCRAFT': 'PIPER',
    'BOEING COMPANY': 'BOEING',
    'BEECH AIRCRAFT CORP': 'BEECH', 'BEECH AIRCRAFT': 'BEECH', 'BEECHCRAFT': 'BEECH'
})

# 3. Count how many times each manufacturer appears
make_counts = cleaned_df['Make'].value_counts()

# 4. Filter to keep only Makes with 50 or more records
valid_makes = make_counts[make_counts >= 50].index
cleaned_df = cleaned_df[cleaned_df['Make'].isin(valid_makes)].copy()

# Print out our remaining rows and the top 10 remaining makes
print(f"Rows remaining after filtering rare makes: {len(cleaned_df)}")
print("\nTop 10 Manufacturers in the cleaned dataset:")
print(cleaned_df['Make'].value_counts().head(10))
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
# 1. Clean up casing, strip whitespaces, and fill any missing models with 'UNKNOWN'
cleaned_df['Model'] = cleaned_df['Model'].str.strip().str.upper()
cleaned_df['Model'] = cleaned_df['Model'].fillna('UNKNOWN')

# 2. Check if models are unique to a single manufacturer
# We'll see if any single model name maps to multiple distinct Makes
model_to_make_counts = cleaned_df.groupby('Model')['Make'].nunique()
non_unique_models = model_to_make_counts[model_to_make_counts > 1]

print(f"Number of model names shared by multiple manufacturers: {len(non_unique_models)}")

# 3. Create a brand new derived column combining Make and Model as a unique identifier
cleaned_df['Make_Model'] = cleaned_df['Make'] + "_" + cleaned_df['Model']

# Take a look at a few examples of our new unique identifier column
print("\nSample of our new unique 'Make_Model' identifiers:")
print(cleaned_df['Make_Model'].head())
```

### Cleaning other columns
- there are other columns containing data that might be related to the outcome of an accident. We list a few here:
- Engine.Type
- Weather.Condition
- Number.of.Engines
- Purpose.of.flight
- Broad.phase.of.flight

Inspect and identify potential cleaning tasks in each of the above columns. Execute those cleaning tasks. 

**Note**: You do not necessarily need to impute or drop NaNs here.

```python
# 1. Clean and standardize string columns (Casing and stripping spaces)
string_cols = ['Engine.Type', 'Weather.Condition', 'Purpose.of.flight', 'Broad.phase.of.flight']

for col in string_cols:
    # Convert to upper case and strip trailing/leading spaces
    cleaned_df[col] = cleaned_df[col].str.strip().str.upper()
    # Fill missing text with a standard 'UNKNOWN' label
    cleaned_df[col] = cleaned_df[col].fillna('UNKNOWN')

# 2. Fix specific messy data quirks
# Standardizing Weather Conditions (sometimes 'IMC'/'VMC' are mixed with 'UNK')
cleaned_df['Weather.Condition'] = cleaned_df['Weather.Condition'].replace({'UNK': 'UNKNOWN'})

# 3. Clean up the 'Number.of.Engines' numerical column
# Fill missing engine counts with the median value so it stays a realistic number
median_engines = cleaned_df['Number.of.Engines'].median()
cleaned_df['Number.of.Engines'] = cleaned_df['Number.of.Engines'].fillna(median_engines)

# Let's peek at a couple of counts to confirm everything looks squeaky clean!
print("Cleaned Weather Conditions:")
print(cleaned_df['Weather.Condition'].value_counts())
print("\nCleaned Broad Phase of Flight:")
print(cleaned_df['Broad.phase.of.flight'].value_counts().head())
```

### Column Removal
- inspect the dataframe and drop any columns that have too many NaNs

```python
# 1. Calculate the percentage of missing values (NaNs) for each column
missing_percentages = cleaned_df.isnull().sum() / len(cleaned_df) * 100

# 2. Display columns that are missing more than 40% of their data
high_missing_cols = missing_percentages[missing_percentages > 40].sort_values(ascending=False)
print("Columns missing more than 40% of data:")
print(high_missing_cols)
```
